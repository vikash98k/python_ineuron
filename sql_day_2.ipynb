Q.7)Write an SQL query to retrieve the top 5 customers with the highest total purchase amount from a table named "Orders" and display their names and total purchase amounts.

```python
import mysql.connector

# Establish a connection to the database
connection = mysql.connector.connect(
    host='localhost',
    user='root',
    password='admin@123',
    database='mysql'
)
# create employee table
cursor = connection.cursor()
sql_command = """CREATE TABLE Orders_record_details (
    ID INT PRIMARY KEY,
    CustomerID INT,
    Amount DECIMAL(10, 2),
    FOREIGN KEY (CustomerID) REFERENCES Customers_record_new_details(ID)
);
"""
cursor.execute(sql_command)
connection.commit()

cursor = connection.cursor()
sql_command = """CREATE TABLE Customers_details (
    ID INT PRIMARY KEY,
    Name VARCHAR(100)
)
"""
cursor.execute(sql_command)
connection.commit()


sql_command = "INSERT INTO Customers_details (ID, Name) VALUES(%s,%s)"
data = [
    (1, 'John Doe'),
    (2, 'Jane Smith'),
    (3, 'Alice Lee'),
    (4, 'Bob Brown')
]
cursor = connection.cursor()
for row in data:
    cursor.execute(sql_command, row)
connection.commit()

data = [
    (1, 1, 500.00),
    (2, 2, 1000.00),
    (3, 3, 750.00),
    (4, 1, 2000.00),
    (5, 2, 1500.00),
    (6, 3, 3000.00),
    (7, 4, 1200.00),
    (8, 4, 800.00),
    (9, 1, 300.00),
    (10, 3, 1800.00)
]
cursor = connection.cursor()
cursor.executemany("INSERT INTO Orders_record (ID, CustomerID, Amount) VALUES(%s,%s,%s)",data)
connection.commit()


sql_command ="""
SELECT c.Name, SUM(o.Amount) AS TotalPurchaseAmount
FROM Customers_record_new_details c
JOIN Orders_record o ON c.ID = o.CustomerID
GROUP BY c.Name
ORDER BY TotalPurchaseAmount DESC
LIMIT 5
"""
cursor = connection.cursor()
cursor.execute(sql_command)
results = cursor.fetchall()
for row in results:
    print(row)

```

Q.8) Write an SQL query to calculate the percentage of sales for each product category in a table named "Sales" and display the category name, total sales amount, and the percentage of total sales.


```python
import mysql.connector

# Establish a connection to the database
connection = mysql.connector.connect(
    host='localhost',
    user='root',
    password='admin@123',
    database='mysql'
)
# create employee table
cursor = connection.cursor()
sql_command = """CREATE TABLE Sales (
    ID INT PRIMARY KEY,
    Category VARCHAR(100),
    Amount DECIMAL(10, 2)
)
"""
cursor.execute(sql_command)
connection.commit()
sql_command = "INSERT INTO Sales (ID, Category, Amount) VALUES(%s,%s,%s)"
data = [
    (1, 'Electronics', 1000.00),
    (2, 'Clothing', 500.00),
    (3, 'Books', 300.00),
    (4, 'Electronics', 1500.00),
    (5, 'Books', 500.00),
    (6, 'Clothing', 800.00),
    (7, 'Electronics', 2500.00),
    (8, 'Books', 700.00),
    (9, 'Clothing', 1200.00),
    (10, 'Electronics', 2000.00)
]
cursor = connection.cursor()

for row in data:
    cursor.execute(sql_command, row)

connection.commit()

sql_command ="""SELECT 
    Category,
    SUM(Amount) AS TotalSalesAmount,
    SUM(Amount) / (SELECT SUM(Amount) FROM Sales) * 100 AS Percentage
FROM Sales
GROUP BY Category;"""
cursor = connection.cursor()
cursor.execute(sql_command)
results = cursor.fetchall()
for row in results:
    print(row)

```

Q.9)Write an SQL query to find the customers who have made the highest total purchases across all years from a table named "Orders" and display their names, email addresses, and the total purchase amount.


```python
import mysql.connector

# Establish a connection to the database
connection = mysql.connector.connect(
    host='localhost',
    user='root',
    password='admin@123',
    database='mysql'
)
# create employee table
cursor = connection.cursor()
sql_command = """CREATE TABLE Customers_record_9 (
    ID INT PRIMARY KEY,
    Name VARCHAR(100),
    Email VARCHAR(100)
)
"""
cursor.execute(sql_command)
connection.commit()
cursor = connection.cursor()
sql_command = """CREATE TABLE Orders_record_9 (
    OrderID INT PRIMARY KEY,
    CustomerID INT,
    Amount DECIMAL(10, 2),
    Year INT,
    FOREIGN KEY (CustomerID) REFERENCES Customers_record_9(ID)
);
"""
cursor.execute(sql_command)
connection.commit()

sql_command = "INSERT INTO Customers_record_9 (ID, Name, Email) VALUES(%s,%s,%s)"
data = [
    (1, 'John Doe', 'johndoe@example.com'),
    (2, 'Jane Smith', 'janesmith@example.com'),
    (3, 'Alice Lee', 'alicelee@example.com')
]
cursor = connection.cursor()
for row in data:
    cursor.execute(sql_command, row)
connection.commit()

cursor = connection.cursor()
data = [
    (1, 1, 1000.00, 2021),
    (2, 2, 1500.00, 2021),
    (3, 3, 1200.00, 2021),
    (4, 1, 800.00, 2022),
    (5, 2, 2000.00, 2022),
    (6, 3, 1800.00, 2022),
    (7, 1, 500.00, 2023),
    (8, 2, 900.00, 2023),
    (9, 3, 1500.00, 2023)
]
cursor.executemany("INSERT INTO Orders_record_9 (OrderID, CustomerID, Amount, Year) VALUES(%s,%s,%s,%s)",data)
connection.commit()

sql_command ="""SELECT
    c.Name,
    c.Email,
    SUM(o.Amount) AS TotalPurchaseAmount
FROM
    Customers_record_9 c
JOIN
    Orders_record_9 o ON c.ID = o.CustomerID
GROUP BY
    c.ID, c.Name, c.Email
HAVING
    SUM(o.Amount) = (
        SELECT
            MAX(TotalAmount)
        FROM
            (SELECT
                CustomerID,
                SUM(Amount) AS TotalAmount
            FROM
                Orders_record_9
            GROUP BY
                CustomerID) AS T
    );
"""
cursor = connection.cursor()
cursor.execute(sql_command)
results = cursor.fetchall()
for row in results:
    print(row)

```
