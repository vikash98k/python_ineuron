```python
import pymysql

# Establish a connection to the database
connection = pymysql.connect(
    host='localhost',
    user='root',
    password='admin@123',
    database='mysql'
)
cursor = connection.cursor()
sql_command = """CREATE TABLE Product_details_table (
    product_id INT PRIMARY KEY,
    name VARCHAR(255),
    stock_quantity INT,
    category_id INT,
    FOREIGN KEY (category_id) REFERENCES category_table(category_id)
)
"""
cursor.execute(sql_command)

# Commit the changes to the database
connection.commit()

# Close the database connection
connection.close()
```

```python
import pymysql

# Establish a connection to the database
connection = pymysql.connect(
    host='localhost',
    user='root',
    password='admin@123',
    database='mysql'
)


sql_command2 = """
    SELECT c.category_name, AVG(p.price) AS average_price
FROM Product_details_table p
JOIN category_details_table c ON p.category_id = c.category_id
GROUP BY c.category_name;

"""
cursor = connection.cursor()
cursor.execute(sql_command2)
results = cursor.fetchall()
for row in results:
    print(row)

```
